# Abstract 

This project examines the relationship between socioeconomic demographics (e.g., educational attainment and racial composition) and the early roll-out of 5G mobile broadband in U.S. cities through the spectrum of education and race. We drew on both FCC broadband deployment data and U.S. census data derived from the American Community Survey. We asked whether communities with a higher proportion of residents that were college educated, white, or wealthier were more likely to have access to 5G. Our exploratory data analysis showed that there was a positive association between education and 5G coverage, while the relationship between race and broadband access was more complicated and uneven. While also having the boundaries of the dataset's granularity and confusion over missing covariates such as affordability and provider-led access, our findings suggest that structural disparities in digital infrastructure might carry over into the 5G space, emphasizing the necessity for equitable planning in future rollouts or changes in technology.

# Research Question

-  Does the distribution of 5G network coverage in major U.S. cities disproportionately favor higher-income, predominantly white, and more educated neighborhoods?
-  This project will explore whether the initial phases of 5G deployment reflect patterns of digital inequity tied to race, income, and education.

Some questions we want to answer: Are lower income neighborhoods less likely to have 5G coverage? How does race or ethnicity correlate with 5G rollout within cities? Are there educational disparities in access to connectivity?

Without access to equitable connectivity, the concerns for digital inequity increase because marginalized groups will not have access to opportunities presented by emerging technologies such as telehealth, smart city services, or advanced mobile applications. This concern is even more troubling given historical data from the Pew Research Center of lower-income, Black, and Hispanic households that disproportionately use mobile internet as their primary connection and may be more reliant on wireless infrastructure.

# Hypothesis


We hypothesize that 5G network coverage is primarily concentrated in wealthier, predominantly white, and higher educated neighborhoods in major U.S. cities. This reflects prior history in which new infrastructure is deployed first to areas with higher economic and commercial value. Lower-income and minority communities seem to have diminished access to the opportunity of 5G more consistently, re-enforcing existing digital divides.

# Data

### Ideal Dataset:

To fully address our research question — whether the rollout of 5G disproportionately favors wealthier, whiter, and more educated neighborhoods — the ideal dataset would include:

- **Granular 5G coverage data** by census tract or ZIP code, from each major carrier (Verizon, AT&T, T-Mobile), including both signal strength and speed.
- **Demographic data** for those same geographic areas, including:
  - Median household income (ACS Table B19013)
  - Racial/ethnic composition (ACS Table B03002)
  - Educational attainment (ACS Table B15003)
- **Geographic boundary files** (e.g., shapefiles) from TIGER/Line to allow spatial joining and mapping of coverage data to demographic data.

Ideally, these datasets would be synchronized in time (e.g., from 2023 or later), geographically aligned, and offer consistent resolution to support tract-level or neighborhood-level analysis within cities. This level of detail would allow us to quantify whether marginalized communities have systematically less access to advanced mobile infrastructure.

### Real Datasets Used:

**Dataset #1: FCC 5G Mobile Broadband Coverage by Census Place**  
- **Source**: [FCC National Broadband Map](https://broadbandmap.fcc.gov/)  
- **File**: `bdc_06_mobile_broadband_summary_by_geography_place_J24_29apr2025.csv`  
- **Content**: Provides modeled 5G coverage percentages (≥7/1 Mbps and ≥35/3 Mbps) for each Census Place in California. This dataset aggregates coverage from multiple providers and is the most detailed publicly available source that can be reliably aligned with ACS data.
- **Limitations**: It does not distinguish between individual providers or include signal quality metrics like latency or tower proximity.

**Dataset #2: ACS Educational Attainment Data (B15003)**  
- **Source**: [U.S. Census Bureau](https://data.census.gov/)  
- **File**: `ACSDT5Y2023.B15003-Data.csv`  
- **Content**: Counts of educational attainment levels for individuals aged 25+ by Census Place. We use this to compute the percent of residents in each place with a Bachelor’s degree or higher.

**Planned Additional Datasets:**
- **Median Income (ACS Table B19013)** — for testing the income disparity dimension of our hypothesis
- **Race/Ethnicity (ACS Table B03002)** — to examine potential racial bias in deployment patterns
- **TIGER/Line Shapefiles** — for enabling geographic joins and spatial visualizations

### Other Feasible Data Sources:
- [FCC National Broadband Map](https://broadbandmap.fcc.gov/) — includes shapefiles and nationwide broadband data by technology and provider
- [T-Mobile & Verizon Coverage Maps](https://www.t-mobile.com/coverage/coverage-map) — useful for visual inspection but limited for quantitative analysis
- [OpenCelliD](https://www.opencellid.org/) — open source tower locations but not complete for all U.S. metros

### Summary of Ideal vs. Real Data:
While our ideal dataset would include provider-level, real-time 5G signal strength at high spatial resolution, the real datasets available through the FCC and Census allow a reasonable and ethically sound approximation. Our approach uses public and statistically valid government sources and avoids questionable methods like screen scraping or relying on proprietary APIs. We align datasets at the **Census Place level**, enabling robust exploratory and statistical analysis of potential digital inequalities in early 5G deployment.

### Merged Dataset:
We created a merged dataset by joining the FCC 5G coverage summary with ACS educational attainment data using a shared geographic identifier (`geography_id`). Each row in the resulting dataset represents a single Census Place in California and includes:
- The percentage of land area covered by 5G at two speed tiers (≥7/1 Mbps and ≥35/3 Mbps)
- The percentage of the adult population (age 25+) holding a bachelor’s degree or higher

This merged dataset enables us to quantitatively analyze the relationship between 5G availability and educational attainment. As additional variables such as income and race/ethnicity are added, this dataset will support correlation, regression, and spatial analyses to assess digital inequity in 5G infrastructure deployment.

## Dataset #1 (use name instead of number here)

```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
import pandas as pd

# === Load FCC 5G Mobile Broadband Coverage Data ===
fcc = pd.read_csv("bdc_06_mobile_broadband_summary_by_geography_place_J24_29apr2025.csv")
fcc = fcc[[
    'geography_id',
    'geography_desc',
    'mobilebb_5g_spd1_area_st_pct',
    'mobilebb_5g_spd2_area_st_pct'
]]
fcc['geography_id'] = fcc['geography_id'].astype(str)
fcc.head()
```

## Dataset #2 (if you have more than one, use name instead of number here)

```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
acs = pd.read_csv("ACSDT5Y2023.B15003-Data.csv")
acs = acs[acs['GEO_ID'] != 'Geography']
acs['geography_id'] = acs['GEO_ID'].str.replace('1600000US', '').str.lstrip('0').astype(str)
edu_cols = ['B15003_001E', 'B15003_022E', 'B15003_023E', 'B15003_024E', 'B15003_025E']
acs[edu_cols] = acs[edu_cols].apply(pd.to_numeric, errors='coerce')
acs['pct_bachelors_or_higher'] = (
    acs['B15003_022E'] + acs['B15003_023E'] +
    acs['B15003_024E'] + acs['B15003_025E']
) / acs['B15003_001E'] * 100
acs = acs[['geography_id', 'NAME', 'pct_bachelors_or_higher']]

## FEEL FREE TO ADD MULTIPLE CELLS PER SECTION 
```

## Merging Dataset #1 and Dataset #2

```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
merged_df = pd.merge(fcc, acs, on='geography_id', how='inner')
merged_df.head()

```

```python
race = pd.read_csv("ACSDT5Y2023.B03002-Data.csv")
race = race[race['GEO_ID'] != 'Geography']
race['geography_id'] = race['GEO_ID'].str.replace('1600000US', '').str.lstrip('0')

race_cols = ['B03002_001E', 'B03002_003E', 'B03002_004E', 'B03002_012E']
race[race_cols] = race[race_cols].apply(pd.to_numeric, errors='coerce')

race['pct_white'] = race['B03002_003E'] / race['B03002_001E'] * 100
race['pct_black'] = race['B03002_004E'] / race['B03002_001E'] * 100
race['pct_hispanic'] = race['B03002_012E'] / race['B03002_001E'] * 100

race = race[['geography_id', 'pct_white', 'pct_black', 'pct_hispanic']]
```

```python
merged_df = pd.merge(merged_df, race, on='geography_id', how='inner')
merged_df.head()
```

```python
merged_df = merged_df.rename(columns={
    'pct_white_x': 'pct_white',
    'pct_black_x': 'pct_black',
    'pct_hispanic_x': 'pct_hispanic'
})
```

# Results

## Exploratory Data Analysis

This exploratory data analysis seeks to investigate the relationship between mobile broadband infrastructure and educational attainment in U.S. census places. By linking institutional information provided by FCC (Federal Communications Commission) regarding coverage and availability of mobile 5G infrastructure with U.S. Census ACS (American Community Survey) documentation of educational attainment information, we examine possible structural inequities in access to digital infrastructure and the socioeconomic contexts that accompany these educational inequities.

## Disparities in 5G Coverage and Educational Attainment Across U.S. Communities



```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.histplot(fcc['mobilebb_5g_spd1_area_st_pct'], ax=axes[0], kde=True)
axes[0].set_title("5G Coverage (Speed Tier 1)")
axes[0].set_xlabel("Percent Area Covered")

sns.histplot(fcc['mobilebb_5g_spd2_area_st_pct'], ax=axes[1], kde=True)
axes[1].set_title("5G Coverage (Speed Tier 2)")
axes[1].set_xlabel("Percent Area Covered")

sns.histplot(acs['pct_bachelors_or_higher'], ax=axes[2], kde=True)
axes[2].set_title("Educational Attainment")
axes[2].set_xlabel("Percent with Bachelor’s or Higher")

plt.tight_layout()
plt.show()
```

**Interpretation:**
- Many census places are represented in full, or close to full in at least one 5G level.
- Educational attainment tends to vary more, such that some locations have less than 10% of the population with a bachelor's degree.
- These unevennesses tend to reveal structural differences in the distribution of education and infrastructure spatially.

## Visualizing Educational Disparities in 5G Speed Tier 2 Coverage



```python
merged = pd.merge(fcc, acs, on='geography_id')
display(merged.describe())
sns.scatterplot(data=merged, x='mobilebb_5g_spd2_area_st_pct', y='pct_bachelors_or_higher')
plt.title("5G Speed 2 Coverage vs. Educational Attainment")
plt.xlabel("5G Coverage (Speed 2, % Area)")
plt.ylabel("% with Bachelor’s or Higher")
plt.grid(True)
plt.show()
```

**Interpretation:**
- We may start to see a loose (and maybe positive) relationship — areas with higher-speed 5G coverage seem to have noticeably higher education levels, while areas with slower 5G, or no coverage often have lower or null levels of education.
- However, the relationship seems to be quite noisy, and other factors (such as higher income, more urbanicity, perhaps more progressive policy) may play a role in this relationship as well.

## Education Levels Across Binned 5G Coverage Categories


```python
merged['coverage_bin'] = pd.cut(
    merged['mobilebb_5g_spd2_area_st_pct'],
    bins=[0, 0.25, 0.5, 0.75, 1.01],
    labels=["Low", "Medium", "High", "Full"]
)

print(merged.groupby('coverage_bin')['pct_bachelors_or_higher'].describe())
sns.boxplot(data=merged, x='coverage_bin', y='pct_bachelors_or_higher')
plt.title("Educational Attainment by 5G Coverage Level")
plt.xlabel("5G Coverage Bin (Speed Tier 2)")
plt.ylabel("% with Bachelor’s or Higher")
plt.grid(True)
plt.show()
```

**Interpretation:**
- In general, education levels tend to increase with higher coverage.
- But there is so much variance in education levels within the bins, that the relationship indicates much more complexity at stake.

## Correlation Between Racial Demographics, Educational Attainment, and 5G Coverage in U.S. Census Regions


```python
heatmap_data = merged_df[[
    'pct_bachelors_or_higher',
    'pct_white',
    'pct_black',
    'pct_hispanic',
    'mobilebb_5g_spd1_area_st_pct',
    'mobilebb_5g_spd2_area_st_pct'
]].dropna()

correlation = heatmap_data.corr()

plt.figure(figsize=(6, 4))
sns.heatmap(
    correlation,
    annot=True,
    cmap='coolwarm',
    fmt=".2f"
)
plt.title("Correlation: Demographics and 5G Coverage")
plt.tight_layout()
plt.show()
```

**Interpretation:**
- We begin to gain a moderate understanding in the relationship — areas with a higher percentage of white residents and college educated residents have more 5G coverage and, specifically, more coverage at higher speeds.
- Alternatively, areas with higher percentages of Hispanic residents show weaker, but still positive correlations with 5G; areas with higher percentages of Black residents have manifested little or no correlations with 5G.
- Nonetheless, the correlation patterns are likely mediating additional factors — such as income level, urban density, or provider infrastructure — that may also help to shape who receives early access to 5G.

# Ethics & Privacy

While our project relies primarily on publicly accessible, aggregated data found with the FCC and U.S. Census Bureau, we feel there is minimal risk to individual privacy, as all data sources, disclosed to us, are in compliance with terms of use and any licensing restrictions. However, ethical considerations extend beyond just privacy, which entails how we frame, analyze, and report our data. Because our research is examining racial, income, and educational disparities in 5G infrastructure application, we need to consider how the data can be misinterpreted or misapplied by others.

Most dangerously, the gaps in coverage may be perceived by policymakers or telecom providers as a sign of weak demand or logistical hurdles, rather than a legacy of structural exclusion over time. This may lead to inadequate investment in those communities who are not great customers for telecom services. To protect against this, we will:

- Clearly articulate our data's limitations such as geographic granularity, coverage thresholds, and lack of provider specific differences.

- Avoid inferring causality where there is only correlation, especially if the demographic variables.

- Use disaggregated subgroup analysis and exploratory stratification to identify potential demographic or geographic biases in 5G Access.

- Frame our findings against historical, structural contexts while avoiding explanations that place downside blame or accountability on individuals or cultures.

- Verify and document any bias on the unaccounted groups or unaffiliated minorities and subgroups in our combined datasets
    
- Add a “Limitations and Ethical Considerations” section to our final report, and be careful with visualizations or summaries to avoid simplification of patterns.

We want to responsibly surface patterns and claims about digital inequality, causing minimal harm, with the intent for our analysis to facilitate rather than reinforce equity-oriented discussions.
